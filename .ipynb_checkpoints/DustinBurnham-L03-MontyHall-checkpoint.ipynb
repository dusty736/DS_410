# 1.  To switch, or not to switch.  That is the question!
The Monty hall problem is an interesting one.  It introduces us to a very unintuitive area of statistics.  Let's set the stage: You're on a game show with three doors.  Behind one door is a car, and a goat behind the other two.  You pick a door, but do not open it.  Basic statisitics tells us that the probability that you picked the car is 1/3.  The host then opens a door that he knows has a goat.  He now presents the choice, do you believe the door you picked has the car or the last door has the car?  This seems like a simple 50-50 proposition considering you are picking between two doors.  But when we consider the idea of conditional probability, the probability of the host's door containing the car GIVEN the fact the other was a goat means there is a 2/3 chance that the car is behind his car.  The only logical answer is switch.  More generally:

N = number of doors

probability that the car is behind your door:

p = 1/N


probability that the car is behind the host's door:

p = (N - 1)/N


## You should definitely switch.  This is unintuitive, and below, I will make the case for why this is true.

```python
# Import libraries
from __future__ import division 

import numpy as np
import pandas as pd
from matplotlib import pyplot
import seaborn as sns
import random
import scipy

# Configure default plot style.
seaborn.set_palette('muted')
seaborn.set_style('ticks')
%matplotlib inline
```

```python
prize = "car"
loser = "goat"

def simulate_mh(door_num, switch):
    """
    Input:
        door_num: the number of doors in this game
        switch: the choice to switch.  True == switching doors at the end.
    Output:
        "win" for car, "loss" for goat.
    """
    
    winner = random.randint(1,door_num)
    keys = list(range(1,door_num+1))
    game = dict(zip(keys,[loser] * door_num))
    game[winner] = prize
    
    choice = game.pop(random.randint(1,door_num))
    
    while(len(game) != 1):
        for key,val in game.items():
            if val == "goat":
                game.pop(key)
                break
    
    if switch == True:
        final_choice = list(game.values())[0]
    else:
        final_choice = choice
        
    if final_choice == "car":
        return("win")
    else:
        return("loss")            
```

```python
def repeat_sims(n, doors, switch):
    """
    Input:
        n: Number of games
        doors: number of doors:
        switch: the choice to switch.  True == switching doors at the end.
    Output:
        Number of wins out of the number of games played.
    """
    switch_win_count = 0

    for i in range(1, n + 1):
        sim = simulate_mh(doors, switch)
        if sim == "win":
            switch_win_count = switch_win_count + 1

    return(switch_win_count)
```

```python
def samp_gen(t, n, doors, switch):
    """
    Input:
        t: Number of samples
        See above function for other args.
    Output:
        win_count: Numpy array containing win counts for t samples.
    """
    
    win_count = np.array([])
    for i in range(t):
        win_count = np.append(win_count, repeat_sims(n, doors, switch))
    return(win_count)
```

```python
# Generate Samples and Calculate mean and expected mean for the case where you switch doors.
doors = np.array([3, 10, 100])
mh_samples = [samp_gen(1000, 100, n, True) for n in doors]

binom_mh_means = [samples.mean() for samples in mh_samples]
expected_mean = 100 * ((doors - 1)/doors)

pd.DataFrame({'Number of Doors':doors, 'Trial mean successes':binom_mh_means,
             'Expected mean successes':expected_mean})
```

```python
# Calculate variance and expected variance
binom_mh_vars = [samples.var() for samples in mh_samples]
expected_vars = 100 * (doors-1)/doors * (1 - ((doors)-1)/doors)

pd.DataFrame({'Number of Doors':doors, 'Trial variance':binom_mh_vars,
             'Expected variance':expected_vars})
```

```python
# Plot histograms and expected distributions for switching 
pyplot.figure(figsize=(20,15))
for i in range(len(doors)):
    pyplot.subplot(len(doors), 1, i+1)
    resolution = 1000
    x_norm = np.linspace(0, 100, resolution)
    y_norm = scipy.stats.norm.pdf(
        x_norm,
        loc=binom_mh_means[i],
        scale=np.sqrt(binom_mh_vars[i]))
    
    pyplot.hist(mh_samples[i], normed=True, color='white', edgecolor='black', label="sample wins")
    pyplot.plot(x_norm, y_norm, color='red', label='Predicted distribution')
    pyplot.xlim(0, 100)
    pyplot.xlabel("Number of winners out of 100")
    pyplot.title(("Histogram of Number of Successes for " + str(doors[i]) + " doors"))
    pyplot.legend()
    
```

```python
# Generate Samples and Calculate mean and expected mean for the case where you keep doors.
doors = np.array([3, 10, 100])
mh_keep_samples = [samp_gen(1000,100,n,False) for n in doors]

binom_mh_keep_means = [samples.mean() for samples in mh_keep_samples]
expected_mean = 100 * (1/doors)

pd.DataFrame({'Number of Doors':doors, 'Mean successes':binom_mh_keep_means,
             'Expected mean successes':expected_mean})
```

```python
# Calculate variance and expected variance
binom_mh_keep_vars = [samples.var() for samples in mh_samples]
expected_keep_vars = 100 * 1/doors * (1 - 1/doors)

pd.DataFrame({'Number of Doors':doors, 'Trial variance':binom_mh_keep_vars,
             'Expected variance':expected_keep_vars})
```

```python
# Plot histograms and expected distributions for switching and keeping for comparing your choices.
pyplot.figure(figsize=(20,15))
for i in range(len(doors)):
    pyplot.subplot(len(doors), 1, i+1)
    resolution = 1000
    x_keep_norm = np.linspace(0, 100, resolution)
    y_keep_norm = scipy.stats.norm.pdf(
        x_keep_norm,
        loc=binom_mh_keep_means[i],
        scale=np.sqrt(binom_mh_keep_vars[i]))
    
    y_norm = scipy.stats.norm.pdf(
        x_norm,
        loc=binom_mh_means[i],
        scale=np.sqrt(binom_mh_vars[i]))
    
    pyplot.hist(mh_keep_samples[i], normed=True, color='m', edgecolor='black', label = 'Keep Sample')
    pyplot.plot(x_norm, y_keep_norm, color='red', label = 'Keep Distribution')
    
    pyplot.hist(mh_samples[i], normed=True, color='b', edgecolor='black', label = 'Switch Sample')
    pyplot.plot(x_norm, y_norm, color='c', label = 'Switch Distribution')
    
    pyplot.xlim(0, 100)
    pyplot.xlabel("Number of winners out of 100")
    pyplot.title(("Histogram of Number of Successes for " + str(doors[i]) + " doors"))
    pyplot.legend()

pyplot.show()
```
